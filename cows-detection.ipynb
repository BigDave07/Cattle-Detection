**STEP 1 CATTLE DETECTION**

```python
# ===== INSTALL AND IMPORT LIBRARIES =====
import io  
import numpy as np
import cv2  # OpenCV - for image processing
import json  # For formatting data as JSON 
from typing import List, Tuple, Dict  # For type hints
from collections import defaultdict  # Special dictionary with default values
from ultralytics import YOLO  # Import YOLO model for object detection

class CattleDetector:
    """
    Cattle detection and tracking system
    Detects cows/calves and assigns unique IDs
    """
    
    def __init__(self, model_name='yolov8n.pt', conf_threshold=0.25):
        """
        Initializes the Detector
        
        Args:
            model_name: YOLO model to use for detection (yolov8n.pt is the smallest/fastest)
            conf_threshold: Confidence threshold for detection (0.25 = 25% sure)
        """
        print(f"Loading Model: {model_name}")
        self.model = YOLO(model_name)  # Load the YOLO AI model
        self.conf_threshold = conf_threshold  # Save threshold for later use
        self.cow_class_id = 19  # "Cow" is class #19 in COCO dataset (what YOLO was trained on)
    
    def detect_cattle(self, image_path: str) -> Dict:
        """
        Detect cattle/calves in a single image
        
        Args:
            image_path: The path to the image file
        
        Returns:
            Dictionary with detection results containing:
            - status: "success" if completed
            - total_detected: number of cattle found
            - detections: list of each animal with position, size, confidence
        """
        # Run Detection - pass image to YOLO model
        results = self.model(image_path, conf=self.conf_threshold)
        r = results[0]  
        
        # Filter for calves/cows only (ignore other objects YOLO might detect)
        detections = []  
        calves_count = 0  # Counter for how many cattle we find
        
        # Loop through every object YOLO detected in the image
        for i, box in enumerate(r.boxes):
            # Convert the class ID from PyTorch tensor to regular Python number
            cls_id = int(box.cls[0].item())
            
            # Process ONLY if it is a calf/cow (class 19)
            if cls_id == self.cow_class_id:
                calves_count += 1  # Increment counter
                
                # Get confidence score 
                conf = float(box.conf[0].item())
                
                # Get bounding box coordinates
                x1, y1, x2, y2 = box.xyxy[0].tolist()
                
                # Calculate center point (needed for checking if calf is isolated)
                center_x = (x1 + x2) / 2
                center_y = (y1 + y2) / 2
                
                # Calculate area of bounding box (width × height)
                # Used to estimate size and compare calves to each other
                area = (x2 - x1) * (y2 - y1)
                
                # Package all info about this calf into a dictionary
                detection = {
                    "id": f"cow_{calves_count}",  # Unique ID (cow_1, cow_2, etc.)
                    "bbox": [int(x1), int(y1), int(x2), int(y2)],  # Box coordinates
                    "center": [int(center_x), int(center_y)],  # Center point
                    "area": int(area),  # Box area in pixels
                    "confidence": round(conf, 3),  # Confidence score (rounded to 3 decimals)
                    "class": "cow"  
                }
                
                # Add this calf to our list of detections
                detections.append(detection)
        
        # Create response package (everything wrapped up nicely for frontend)
        response = {
            "status": "success",  # Status flag
            "image_path": image_path,  # Which image we analyzed
            "total_detected": calves_count,  # How many cattle found
            "detections": detections,  # List of all individual cattle with their info
            "model_info": {  # Info about the model we used
                "model": "yolov8n",
                "confidence_threshold": self.conf_threshold
            }
        }
        
        return response  # Send back all the results
    
    def get_annotated_image(self, image_path: str) -> np.ndarray:
        """
        Get image with bounding boxes drawn around detected cattle
        
        Args:
            image_path: The path to the image file
        
        Returns:
            Annotated image as numpy array (grid of RGB pixel values)
            - Has colored boxes drawn around each detected animal
            - Includes labels with class name and confidence
        """
        # Run detection on the image
        results = self.model(image_path, conf=self.conf_threshold)
        
        # Use YOLO's built-in plotting function to draw boxes
        # This automatically adds colored rectangles and labels
        annotated = results[0].plot()
        
        return annotated  # Return the image with boxes drawn on it


# ===== TEST THE DETECTOR =====
print("=" * 50)
print("CATTLE DETECTOR INITIALIZED")
print("=" * 50)

# Create an instance of our detector (loads YOLO model)
detector = CattleDetector(conf_threshold=0.25)
print("✓ Detector ready!\n")
```

**Step 2: Health Analyzer**

```python
# ===== HEALTH ANALYSIS MODULE (V1) =====
import math
from typing import List, Dict

class HealthAnalyzer:
    """
    Rule-based health analyzer for calves/cattle.

    Uses only what YOLO gives you:
      - bbox (position + width/height)
      - center
      - area

    Checks:
      - Is calf smaller than others?  -> 'undersized'
      - Is calf far from others?      -> 'isolated'
      - Is calf near image edges?     -> 'near_edge'
      - Rough posture (lying/standing)-> 'lying_down'

    (Later you can add ears/eyes/tail/rib models on top of this.)
      - We can plug in extra visual features like:
        * ear_droop_score
        * eye_clarity_score
        * tail_dirt_score
        * rib_visibility_score
        * rate_of_gain, movement_stats (from time-series)
    """
    
    def __init__(
        self,
        isolation_threshold: float = 200.0,
        size_variation_threshold: float = 0.6,
        wide_to_tall_ratio_for_lying: float = 1.6
    ):
        """
        Args:
            isolation_threshold:
                If nearest neighbor is farther than this (pixels),
                we mark calf as 'isolated'.
            size_variation_threshold:
                If calf area < threshold * average_area,
                we mark it as 'undersized'.
            wide_to_tall_ratio_for_lying:
                Simple posture rule:
                  width / height > this  -> lying
                  else                   -> standing
        """
        self.isolation_threshold = isolation_threshold
        self.size_variation_threshold = size_variation_threshold
        self.wide_to_tall_ratio_for_lying = wide_to_tall_ratio_for_lying
    
    # ---------- SMALL HELPER FUNCTIONS ----------
    def calculate_distance(self, point1: List[int], point2: List[int]) -> float:
        """
        Distance between two points using Pythagoras:
        d = sqrt((x2-x1)^2 + (y2-y1)^2)
        """
        x1, y1 = point1
        x2, y2 = point2
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
    def find_nearest_neighbor(self, cow_id: str, detections: List[Dict]) -> Dict:
        """
        For a given calf ID, find the closest other calf.
        Used to decide if it is isolated.
        """
        # Find the current calf
        current_cow = None
        for cow in detections:
            if cow["id"] == cow_id:
                current_cow = cow
                break
        
        # If it's missing or the only one in the image
        if not current_cow or len(detections) == 1:
            return {
                "nearest_distance": None,
                "nearest_id": None,
                "is_isolated": len(detections) == 1  # True if it is completely alone
            }
        
        min_distance = float("inf")
        nearest_id = None
        
        # Compare to every other calf
        for cow in detections:
            if cow["id"] == cow_id:
                continue  # skip itself
            
            dist = self.calculate_distance(current_cow["center"], cow["center"])
            
            if dist < min_distance:
                min_distance = dist
                nearest_id = cow["id"]
        
        is_isolated = min_distance > self.isolation_threshold
        
        return {
            "nearest_distance": round(min_distance, 2),
            "nearest_id": nearest_id,
            "is_isolated": is_isolated
        }
    
    def analyze_size(self, detections: List[Dict]) -> Dict:
        """
        Compare each calf's area with the average area in the image.
        Smaller than (threshold * average) → 'undersized'.
        """
        if not detections:
            return {}
        
        areas = [cow["area"] for cow in detections]
        avg_area = sum(areas) / len(areas)
        
        size_analysis = {}
        
        for cow in detections:
            cow_id = cow["id"]
            area = cow["area"]
            
            size_ratio = area / avg_area  # 1.0 = same as average
            is_undersized = size_ratio < self.size_variation_threshold
            
            size_analysis[cow_id] = {
                "area": area,
                "size_ratio": round(size_ratio, 2),
                "is_undersized": is_undersized,
                "status": "small" if is_undersized else "normal"
            }
        
        return size_analysis
    
    def check_edge_position(
        self,
        cow: Dict,
        image_width: int,
        image_height: int,
        margin: int = 50
    ) -> bool:
        """
        Check if calf's center is within 'margin' pixels of any edge.
        """
        x, y = cow["center"]
        
        near_left = x < margin
        near_right = x > (image_width - margin)
        near_top = y < margin
        near_bottom = y > (image_height - margin)
        
        return near_left or near_right or near_top or near_bottom
    
    def estimate_posture(self, cow: Dict) -> str:
        """
        Rough guess: lying or standing, based on bbox shape.

        width / height:
          - very wide  -> likely lying
          - more square/tall -> likely standing
        """
        x1, y1, x2, y2 = cow["bbox"]
        width = x2 - x1
        height = y2 - y1
        
        if height == 0:
            return "unknown"
        
        ratio = width / height
        
        if ratio > self.wide_to_tall_ratio_for_lying:
            return "lying"
        else:
            return "standing"
    
    # ---------- MAIN HEALTH ANALYSIS ----------
    def analyze_health(
        self,
        detections: List[Dict],
        image_width: int,
        image_height: int,
        visual_features: Dict = None 
    ) -> Dict:
        """
        Run full health analysis on ONE image.

        Input:
          - detections: list from CattleDetector.detect_cattle()
          - image_width, image_height: for edge checks

        Output:
          JSON-friendly dict with:
            - total_analyzed
            - healthy
            - flagged
            - health_data: list with per-calf info
        """
        if not detections:
            return {
                "total_analyzed": 0,
                "healthy": 0,
                "flagged": 0,
                "health_data": [],
                "analysis_parameters": {
                    "isolation_threshold": self.isolation_threshold,
                    "size_variation_threshold": self.size_variation_threshold,
                    "wide_to_tall_ratio_for_lying": self.wide_to_tall_ratio_for_lying
                }
            }
        
        # First, analyze sizes once (needs all calves to compute average)
        size_analysis = self.analyze_size(detections)
        
        health_data = []
        flagged_count = 0
        
        for cow in detections:
            cow_id = cow["id"]
            
            # 1) Isolation
            neighbor_info = self.find_nearest_neighbor(cow_id, detections)
            
            # 2) Edge position
            at_edge = self.check_edge_position(cow, image_width, image_height)
            
            # 3) Posture
            posture = self.estimate_posture(cow)
            
            # 4) Size
            size_info = size_analysis[cow_id]
            
            # Collect flags (reasons to worry)
            flags = []
            if neighbor_info["is_isolated"]:
                flags.append("isolated")
            if size_info["is_undersized"]:
                flags.append("undersized")
            if at_edge:
                flags.append("near_edge")
            if posture == "lying":
                flags.append("lying_down")
            
            # Final status
            if flags:
                health_status = "attention_needed"
                flagged_count += 1
            else:
                health_status = "healthy"
            
            cow_health = {
                "id": cow_id,
                "bbox": cow["bbox"],
                "center": cow["center"],
                "confidence": cow["confidence"],
                "health_status": health_status,
                "flags": flags,
                "metrics": {
                    "nearest_distance": neighbor_info["nearest_distance"],
                    "nearest_neighbor": neighbor_info["nearest_id"],
                    "size_ratio": size_info["size_ratio"],
                    "at_edge": at_edge,
                    "posture": posture
                }
            }
            
            health_data.append(cow_health)
        
        return {
            "total_analyzed": len(detections),
            "healthy": len(detections) - flagged_count,
            "flagged": flagged_count,
            "health_data": health_data,
            "analysis_parameters": {
                "isolation_threshold": self.isolation_threshold,
                "size_variation_threshold": self.size_variation_threshold,
                "wide_to_tall_ratio_for_lying": self.wide_to_tall_ratio_for_lying
            }
        }
```

```python
# ===== QUICK TEST: DETECTION + HEALTH ANALYSIS + VIS =====

import cv2
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 1) Create detector & analyzer
print("=" * 60)
print("INITIALIZING MODELS")
print("=" * 60)

detector = CattleDetector()
analyzer = HealthAnalyzer(                     
    isolation_threshold=200.0,      # how far = isolated (pixels)
    size_variation_threshold=0.6,   # <60% of average size = undersized
    wide_to_tall_ratio_for_lying=1.6
)

print("✓ CattleDetector and HealthAnalyzer ready!\n")

# 2) Choose test image
img_path = "/kaggle/input/test-images/testimg2.jpg"  # change if needed
print(f"Analyzing image: {img_path}\n")

# 3) Run detection
detection_results = detector.detect_cattle(img_path)
print(f"✓ Detected {detection_results['total_detected']} cattle\n")

# 4) Get image size (needed for edge checks)
img_bgr = cv2.imread(img_path)
if img_bgr is None:
    raise FileNotFoundError(f"Could not read image at {img_path}")

img_height, img_width = img_bgr.shape[:2]
print(f"Image size: {img_width} x {img_height} pixels\n")

# 5) Run health analysis
health_report = analyzer.analyze_health(
    detections=detection_results["detections"],
    image_width=img_width,
    image_height=img_height
)

print("=" * 60)
print("HEALTH ANALYSIS REPORT (JSON)")
print("=" * 60)
print(json.dumps(health_report, indent=2))

# 6) Visualization (draw boxes with health status on image)
print("\nCreating visualization...")

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, figsize=(14, 10))
ax.imshow(img_rgb)

for cow in health_report["health_data"]:
    x1, y1, x2, y2 = cow["bbox"]
    
    # color depends on health status
    if cow["health_status"] == "healthy":
        color = "green"
        label_bg = "lightgreen"
    else:
        color = "red"
        label_bg = "lightcoral"
    
    # draw bounding box
    rect = patches.Rectangle(
        (x1, y1),
        x2 - x1,
        y2 - y1,
        linewidth=3,
        edgecolor=color,
        facecolor="none"
    )
    ax.add_patch(rect)
    
    # build label text: ID + status + any flags
    label = f"{cow['id']} | {cow['health_status']}"
    if cow["flags"]:
        label += "\n" + ", ".join(cow["flags"])
    
    # draw label above the box
    ax.text(
        x1,
        max(y1 - 10, 0),
        label,
        bbox=dict(boxstyle="round", facecolor=label_bg, alpha=0.8),
        fontsize=10,
        color="black",
        weight="bold"
    )

ax.axis("off")
title = f"Health Analysis: {health_report['healthy']} Healthy, {health_report['flagged']} Flagged"
ax.set_title(title, fontsize=18, weight="bold", pad=20)
plt.tight_layout()
plt.show()

print("\n✓ QUICK TEST COMPLETE!")
```
